```python
import h5py
## reduce H5 file up to 100MB for github

def select_and_save_data(input_file, output_file, selected_dataset_name):
    # Read the HDF5 file
    with h5py.File(input_file, 'r') as file:
        # Check if the selected dataset exists in the file
        if selected_dataset_name not in file:
            print(f"The dataset '{selected_dataset_name}' does not exist in the file.")
            return

        # Select the dataset you want to keep
        selected_data = file[selected_dataset_name][:268]

    # Save the selected data to a new HDF5 file
    with h5py.File(output_file, 'w') as new_file:
        new_file.create_dataset(selected_dataset_name, data=selected_data)


input_file_name = "SERVER/backend/VM_data.h5"
output_file_name = "SERVER/backend/VM_data_red.h5"
dataset_to_keep = "strain"

# Call the function
select_and_save_data(input_file_name, output_file_name, dataset_to_keep)
```
